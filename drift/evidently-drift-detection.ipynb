```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import evidently
```

```python
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset
```

```python
# Grab the data
wine_data = load_wine(as_frame=True)
feature_names = wine_data.feature_names
X, y = wine_data.data, wine_data.target 

# Make a 50/50 reference/test split
X_ref, X_test, y_ref, y_test = train_test_split(X, y,
                                                test_size=0.50,
                                                random_state=42)
```

```python
# data_drift_report = Report(metrics=[
#     DataDriftPreset(num_stattest='ks', cat_stattest='psi', num_stattest_threshold=0.2, cat_stattest_threshold=0.2),
# ])
# colums_for_report = ['alcohol', 'malic_acid']
data_drift_report = Report(metrics = [DataDriftPreset()])
# data_drift_report.run(reference_data=X_ref[colums_for_report], current_data=X_test[colums_for_report])
data_drift_report.run(reference_data=X_ref, current_data=X_test)
```
